```python
from selenium import webdriver
import pandas as pd
import time

#Lists for project programs
names_of_programs = []
names_of_instutions = []
status_of_programs = []
deadlines_of_programs = []

#Options for selenium driver
op = webdriver.ChromeOptions()
PATH = "C:\Program Files (x86)\chromedriver.exe"
op.add_argument("--headless")

#Opening driver and conecting to the website
yatirimadestek_driver = webdriver.Chrome(PATH, chrome_options=op)
yatirimadestek_driver.get("https://www.yatirimadestek.gov.tr/gelismis-arama")

#Finding the list of projects from the website
project_list = yatirimadestek_driver.find_element_by_class_name("arama-sonuclar")
projects = project_list.find_elements_by_class_name("item")

#Taking informations from every one of project program
for project in projects:
    #Taking the name of instution which provides the budget for the program
    instution = project.find_element_by_class_name("icol.col1")
    #Taking the name of the project program
    program = project.find_element_by_class_name("icol.col2").text
    #Taking the activity status of the program
    statute = project.find_element_by_class_name("icol.col3").text
    #If status of the program is 'Active' then we need rows below to take the deadline information of the program
    if statute == "AKTİF":
        #Clicking to object of the list to see program details
        project.click()
        #Taking details of the program
        active_project_details = project.find_element_by_class_name("detail-props")
        #Taking the information of deadline
        date = active_project_details.find_elements_by_class_name("dpval")[2].text
        #printing project details and appending them to the lists which are defined at the start of algorithm
        print(f'Kurum Adı:{instution.text} | Program Adı:{program} | Çağrı Durumu:{statute} | Son Başvuru Tarihi:{date}')
        names_of_programs.append(program)
        names_of_instutions.append(instution.text)
        status_of_programs.append(statute)
        deadlines_of_programs.append(date)
    #If the status of program is 'Deactive' then we need rows below to take the deadline as '-'
    else:
        print(f'Kurum Adı:{instution.text} | Program Adı:{program} | Çağrı Durumu:{statute}')
        names_of_programs.append(program)
        names_of_instutions.append(instution.text)
        status_of_programs.append(statute)
        deadlines_of_programs.append("-")

#Closing the driver
yatirimadestek_driver.close()


    
```

```python
#Options for selenium driver
op = webdriver.ChromeOptions()
PATH = "C:\Program Files (x86)\chromedriver.exe"
op.add_argument("--headless")

#Opening driver and conecting to the website
tubitak_ua_driver = webdriver.Chrome(PATH, chrome_options=op)
tubitak_ua_driver.get("https://uidb-pbs.tubitak.gov.tr/#tabs-3")

#Finding the list of projects from the website
program_table = tubitak_ua_driver.find_elements_by_xpath('//*[@id="programCagriListTable"]/tbody')[1]
project_list = program_table.find_elements_by_tag_name("tr")

#Taking informations from every one of project program
for project in project_list:
    #Taking the name of the project program
    program = project.find_elements_by_tag_name("td")[3].text
    #Taking the name of instution which provides the budget for the program (For this website every instution which provides budget for the programs is 'TUBİTAK')
    instution = 'TUBİTAK'
    #Taking the activity status of the program (For this website all of the programs is 'Active'('Aktif' in turkish))
    statute = 'AKTİF'
    #Taking deadline information of the program
    date = project.find_elements_by_tag_name("td")[6].text

    #printing project details and appending them to the lists which are defined at the start of algorithm
    print(f'Kurum Adı:{instution} | Program Adı:{program} | Çağrı Durumu:{statute} | Son Başvuru Tarihi:{date}')
    names_of_programs.append(program)
    names_of_instutions.append(instution)
    status_of_programs.append(statute)
    deadlines_of_programs.append(date)

#Closing the driver
tubitak_ua_driver.close() 



```

```python
#Options for selenium driver
op = webdriver.ChromeOptions()
PATH = "C:\Program Files (x86)\chromedriver.exe"
op.add_argument("--headless")

#Opening driver and conecting to the website
ab_driver = webdriver.Chrome(PATH, chrome_options=op)
ab_driver.get("https://hibeler.ab.gov.tr/#/grants")

time.sleep(2)

#Finding the list of projects from the website
ab_project_table = ab_driver.find_element_by_xpath('//*[@id="kt_content"]/div/div/kt-grantlistforuser/div/div[2]/div[2]/table')
project_list = ab_project_table.find_elements_by_class_name('ng-star-inserted')

#Taking the names of the project programs
program_names = ab_project_table.find_elements_by_xpath('//*[@class="ng-star-inserted"]/td[2]/a')
#Taking the names of instutions which provides the budget for the program
instutions = ab_project_table.find_elements_by_xpath('//*[@class="ng-star-inserted"]/td[2]/div/span')
#Taking the information of deadlines
dates = ab_project_table.find_elements_by_xpath('//*[@class="ng-star-inserted"]/td[5]/span')

#Taking the name every one of project program
for program_name in program_names:
    print(program_name.text)
    names_of_programs.append(program_name.text)
    #Taking the activity status of the program (For this website all of the programs is 'Active'('Aktif' in turkish))
    status_of_programs.append('AKTİF')
#Taking the name every one of instution
for instution in instutions:
    print(instution.text)
    names_of_instutions.append(instution.text)
#Taking the deadline every one of project program
for date in dates:
    print(date.text)
    deadlines_of_programs.append(date.text)

ab_driver.close()
```

```python
#Making a dict from the lists which are defined at the start of the algorithm
cagrilar = {"Progra Adı": names_of_programs, "Kurum Adı": names_of_instutions, "Çağrı Durumu": status_of_programs, "Son Başvuru Tarihi": deadlines_of_programs}
cagrilar_df = pd.DataFrame(cagrilar)
cagrilar_df = cagrilar_df.sort_values('Kurum Adı')
cagrilar_df
```
